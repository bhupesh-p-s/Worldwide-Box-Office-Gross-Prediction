## Data Pre-Processing

#### Formating the Box Office Column to float

```python
df1['description']=df1['description'].str.replace('$','')
df1['description']=df1['description'].str.replace(',','')
df1['description']=df1['description'].astype(float)
```

```python
df1["description"] = df1["description"] / 1e9
df1["description"] = df1["description"].round(3).astype(float)
```

#### Generating some stats on genre

```python
df2.genre = df2.genre.apply(lambda x: x.strip())
genre_stats = df2['genre'].value_counts(ascending=False)
genre_stats
```

```python
genre_stats_less_than_10 = genre_stats[genre_stats<=10]
genre_stats_less_than_10
```

```python
df3 = df2.copy()
```

#### Classifying genres which is less than 10 as 'others'

```python
df3.genre = df2.genre.apply(lambda x: 'other' if x in genre_stats_less_than_10 else x)
len(df3.genre.unique())
```

#### Dropping unnecessary columns

```python
x = df3.drop(['name','star_3','star_4','description'],axis='columns')
```

## Building Dependent and Independent variables

#### Performing one hot encoding

```python
X = pd.get_dummies(x)
```

```python
X.head()
```

```python
X=X.drop(["genre_other"],axis="columns")
```

```python
y = df3.description
```

```python
y.tail()
```

## Model Building

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
rfg = RandomForestRegressor()
rfg.fit(X_train,y_train)
rfg.score(X_test,y_test)
```

```python
grd = GradientBoostingRegressor()
grd.fit(X_train,y_train)
grd.score(X_test,y_test)
```

```python
svr = SVR()
svr.fit(X_train,y_train)
svr.score(X_test,y_test)
```

#### Trying Neural Network Regressor

```python
import tensorflow as tf
```

```python
# Create a sequential model
model = tf.keras.models.Sequential()

# Add a fully connected layer with 128 units and ReLU activation
model.add(tf.keras.layers.Dense(128, activation='relu'))

# Add another fully connected layer with 128 units and ReLU activation
model.add(tf.keras.layers.Dense(128, activation='relu'))

# Add an output layer with a single unit
model.add(tf.keras.layers.Dense(1))
```

```python
# Compile the model with the Mean Squared Error loss function and the Adam optimizer
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Train the model using the training data
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
test_loss = model.evaluate(X_test,y_test)
print('Test loss:', test_loss)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = model.predict(X_test)

# Calculate the baseline MSE by assuming the mean of the target variable is the prediction for all instances
baseline_mse = mean_squared_error(y_test, [y_train.mean() for i in range(len(y_test))])

# Calculate the model's MSE
model_mse = mean_squared_error(y_test, y_pred)

# Calculate the gain
gain = (baseline_mse - model_mse) / baseline_mse * 100

print("Gain: {:.2f}%".format(gain))
```

## Predicting the Future Box Office

```python
def predict_price(genre, director, star_1, star_2):
    X_new = pd.DataFrame({'genre': genre, 'director': director, 'star_1': star_1, 'star_2': star_2}, index=[0])
    X_new_encoded = pd.get_dummies(X_new)
    
    # Ensure that the new data has the same columns as the training data
    missing_cols = set(X.columns) - set(X_new_encoded.columns)
    for col in missing_cols:
        X_new_encoded[col] = 0
    X_new_encoded = X_new_encoded[X.columns]
    
    return grd.predict(X_new_encoded)[0]
```
