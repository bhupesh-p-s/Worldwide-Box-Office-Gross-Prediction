```python
def selection_sort(arr):
    sorted = []
    for i in range(0,len(arr)):
        minimum = arr.pop(arr.index(min(arr)))
        sorted.append(minimum)
    return sorted
```

```python
selection_sort([3,4,5,1,2])
```

```python
def bubble_sort(arr):
    sorted = []
    for i in range(0,len(arr)-1):
        for j in range(0,len(arr)-i-1):
            if arr[j]>arr[j+1]:
                swap = arr[j+1]
                arr[j+1] = arr[j]
                arr[j] = swap
            
    return arr
```

```python
bubble_sort([3,4,5,1,2])
```

```python
def insertion_sort(arr):
    sorted = []
    for i in range(1,len(arr)-1):
        if arr[i-1]>arr[i]:
            swap = arr[i-1]
            arr[i-1]=arr[i]
            arr[i] = swap
        elif arr[i]<arr[i+1]:
    return arr
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.imdb.com/list/ls098063263/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

movies = []
counter = 0

while counter < 100:
    for movie in soup.find_all("div", class_="lister-item mode-detail"):
        movie_info = {}
        movie_info["name"] = movie.h3.a.text
        movie_info["genre"] = movie.find("span", class_="genre").text.strip()
        crew = []
        for tag in movie.find_all("p", class_="text-muted text-small"):
            for link in tag.find_all("a"):
                crew.append(link.text)
        if crew:
            movie_info["director"] = crew[0]
        if len(crew) > 1:
            movie_info["star_1"] = crew[1]
        if len(crew) > 2:
            movie_info["star_2"] = crew[2]
        if len(crew) > 3:
            movie_info["star_3"] = crew[3]
        if len(crew) > 4:
            movie_info["star_4"] = crew[4]
        description = movie.find("div", class_="list-description").p.b.text
        movie_info["description"] = description
        movies.append(movie_info)

    next_page = soup.find("a", class_="flat-button lister-page-next next-page")
    if not next_page:
        break

    response = requests.get("https://www.imdb.com" + next_page["href"])
    soup = BeautifulSoup(response.text, "html.parser")
    counter+=1



```

```python
df = pd.DataFrame(movies)
df.to_csv("movies.csv",index=False)
```

```python
def predict_price(genre, director, star_1, star_2):
    X_new = pd.DataFrame({'genre': genre, 'director': director, 'star_1': star_1, 'star_2': star_2}, index=[0])
    X_new_encoded = pd.get_dummies(X_new)
    
    # Ensure that the new data has the same columns as the training data
    missing_cols = set(X.columns) - set(X_new_encoded.columns)
    for col in missing_cols:
        X_new_encoded[col] = 0
    X_new_encoded = X_new_encoded[X.columns]
    
    return grd.predict(X_new_encoded)[0]
```

```python
#Example input values for predict_price function
genre = 'Action, Adventure, Fantasy'
director = 'Quentin Tarantino'
star_1 = 'Leonardo Dicaprio'
star_2 = 'Bradd Pitt'

# Call the predict_price function
predicted_price = predict_price(genre,director,star_1,star_2)

# Output the predicted price
print("Predicted Box Office Price:",predicted_price)
```

```python
import tensorflow as tf
```

```python
# Create a sequential model
model = tf.keras.models.Sequential()

# Add a fully connected layer with 128 units and ReLU activation
model.add(tf.keras.layers.Dense(128, activation='relu'))

# Add another fully connected layer with 128 units and ReLU activation
model.add(tf.keras.layers.Dense(128, activation='relu'))

# Add an output layer with a single unit
model.add(tf.keras.layers.Dense(1))
```

```python
# Compile the model with the Mean Squared Error loss function and the Adam optimizer
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Train the model using the training data
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
test_loss = model.evaluate(X_test,y_test)
print('Test loss:', test_loss)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = model.predict(X_test)

# Calculate the baseline MSE by assuming the mean of the target variable is the prediction for all instances
baseline_mse = mean_squared_error(y_test, [y_train.mean() for i in range(len(y_test))])

# Calculate the model's MSE
model_mse = mean_squared_error(y_test, y_pred)

# Calculate the gain
gain = (baseline_mse - model_mse) / baseline_mse * 100

print("Gain: {:.2f}%".format(gain))
```

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
import pandas as pd

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'random_forest': {
            'model': RandomForestRegressor(),
            'params': {
                'n_estimators': [50, 100, 200],
                'criterion': ['squared_error', 'absolute_error']
            }
        },
        'gradient_boosting': {
            'model': GradientBoostingRegressor(),
            'params': {
                'n_estimators': [50, 100, 200],
                'learning_rate': [0.1, 0.05, 0.01],
                'loss': ['ls', 'lad', 'huber', 'quantile']
            }
        },
        'svr': {
            'model': SVR(),
            'params': {
                'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
                'C': [0.1, 1, 10, 100, 1000],
                'degree': [2, 3, 4, 5]
            }
        },
    
        
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(X,y)

```
